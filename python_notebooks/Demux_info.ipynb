```python
import pandas as pd, numpy as np, itertools
```

```python
my_info = pd.read_csv('my_info.tsv', sep = "\t", header=[0, 1])
```

```python
hash_info = pd.read_csv('hashing_info.tsv', sep='\t')
```

```python
idx = pd.IndexSlice
```

```python
hash_info
```

```python
my_info
```

```python
result = pd.DataFrame({'Demux_method':['Calico_solo', 'MULTIseq', 'HTOdemux']*my_info.shape[0], 'Samples':list(itertools.chain.from_iterable(itertools.repeat(x, 3) for x in my_info['LAB']['SAMPLE']))})
```

```python
result["HTO1"] = ""
result["HTO2"] = ""
result["HTO3"] = ""
result["HTO4"] = ""
result["HTO5"] = ""
result["HTO6"] = ""
result["HTO7"] = ""
result["HTO8"] = ""
result["HTO9"] = ""
result["HTO10"] = ""
```

```python
result
```

```python
my_info.loc[my_info.loc[:, idx['LAB', 'SAMPLE']] == "NPSAD-20201112-A2-cDNA", idx['STARsolo', 'N_CELLS_AFTER_DEMUX_MS']].values[0]
```

```python
my_info.loc[my_info.loc[:, idx['LAB', 'SAMPLE']] == "NPSAD-20201013-A1-cDNA", idx['STARsolo', 'N_CELLS_AFTER_DEMUX_MS']].values[0].split(',')
```

```python
for x in zip(result['Demux_method'].to_list(), result['Samples'].to_list()):
    if x[0].startswith('C'):
        vals = my_info.loc[my_info.loc[:, idx['LAB', 'SAMPLE']] == x[1], idx['STARsolo', 'N_CELLS_AFTER_DEMUX_CS']].values[0].split(',')
    elif x[0].startswith('M'):
        vals = my_info.loc[my_info.loc[:, idx['LAB', 'SAMPLE']] == x[1], idx['STARsolo', 'N_CELLS_AFTER_DEMUX_MS']].values[0].split(',')
    else:
        vals = my_info.loc[my_info.loc[:, idx['LAB', 'SAMPLE']] == x[1], idx['STARsolo', 'N_CELLS_AFTER_DEMUX_HD']].values[0].split(',')
        
    for v in vals:
        sub = v.split(':')[0]
        n_cells = v.split(':')[1]
        hto = hash_info.loc[(hash_info['cDNA_ID'] == x[1]) & (hash_info['SubID'] == sub), 'hashtag'].values[0]
        result.loc[(result['Demux_method'] == x[0]) & (result['Samples'] == x[1]), hto] = n_cells
    
```
